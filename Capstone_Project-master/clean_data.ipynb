## Cause of Loss Reformat

```python
df['Normalized Cause of Loss'] = df['Cause of Loss'].apply(lambda x: str(x).lower())
```

```python
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'fire\w*', '3', str(x)))
#Explosion:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'explo\w+|blast|vapor cloud|jet fire', '4', str(x)))
#Toxic Release:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'toxic|health', '5', str(x)))
#Lightning:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'lightning|lightning strike', '6', str(x)))
#Blackout:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'power|outage|blackout', '7', str(x)))
#Technical Failure:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'machine|breakdown|pump|compressor|turbine|blade|heater|boiler|technical failure', '8', str(x)))
#Pipeline Ruptures:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'pipe\w+|rupt\w+|breakage', '9', str(x)))
#Leaks and Spills:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'lea\w+|spil\w+|overflow|release', '10', str(x)))
#Derailment:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'train|derail|rail', '11', str(x)))
#Armed Attacks:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'terror|attack|weapon', '12', str(x)))
#Bombing:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'bomb', '13', str(x)))
#Kidnapping:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'kidnap|official|director|officer', '14', str(x)))
#Sabotage:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'sabotage', '15', str(x)))

#Cyber Attack:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'breach|security|internet|online', '16', str(x)))
#Protests and Blockades:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'protest|vandalism|riot', '17', str(x)))

#Strikes:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'strike', '18', str(x)))

#Allegations:

#Work Place Injuries:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'worker compensation|inju\w+|human|death\w+|fatali\w+', '20', str(x)))

#maintenance:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'maintenance|machine|breakdown|pump|compressor|turbine|blade|heater|boiler', '21', str(x)))

#Regulatory Changes:

#Natural Disasters:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'weather|hurricane|storm|wind|flood|tornado|storm surge|other', '23', str(x)))

#Hazardous Contamination:
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'contamination|hazard|soil|pollution|emission\w+', '24', str(x)))

df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'[a-z]|[A-Z]|[\-()/\,\s]', ' ', str(x)))

df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.sub(r'\s+', ' ', str(x)))
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: re.findall(r'[0-9]+', str(x)))
df['Normalized Cause of Loss'] = df['Normalized Cause of Loss'].apply(lambda x: str(x).replace("[","").replace("]","").replace("'","").replace(" ",""))
```

## Restructure Loss Estimated (Gallons|Barrels|Pounds)

```python
df['Loss Estimated (Gallons|Barrels|Pounds)'] = df['Loss Estimated (Gallons|Barrels|Pounds)'].fillna(0)
```

```python
df['Loss Estimated (Gallons|Barrels|Pounds)'][df['Loss Estimated (Gallons|Barrels|Pounds)'] != 0].tail(10)
```

```python
def extract_unit_value(df, new_column, unit):
    '''
    Extract value with specific unit from a potential units: gallons, barrels, pounds
    and columnize each unit
    '''
    df[new_column] = df['Loss Estimated (Gallons|Barrels|Pounds)']
    #collect the right value with right unit
    df[new_column] = df[new_column].apply(lambda x: re.findall(r'[0-9,]+(?=\s' + unit + ')', str(x)))
    #replace empty list by 0
    df[new_column] = df[new_column].apply(lambda x: re.sub(r'\[\]', "0", str(x)))
    #choose first element in list of possible values
    df[new_column] = df[new_column].apply(lambda x: re.findall(r'[0-9,]+', str(x))[0])
    #remove comma 
    df[new_column] = df[new_column].apply(lambda x: int(re.sub(r',', "", str(x))))
    #convert number to integer
    df[new_column] = df[new_column].astype(int)
    return df
```

```python

#create new column with specified unit
df = extract_unit_value(df, 'Gallons', 'gallons')
```

```python
import pandas as pd
import re
```

```python
file = "sources_from_links/accident_hazardous_liquid_1986_jan2002/accident_hazardous_liquid_1986_jan2002.csv"
```

```python
data = pd.read_csv(file)
```

```python
df = pd.DataFrame(data=data, columns=[ 'IDATE','ACSTATE', 'CAUS','NAME', 'NARRATIVE', 'LOSS'])
```

```python
# # df.NARRATIVE.values.any
# empty_list = []
# df.NARRATIVE.fillna('').head()
# # df.NARRATIVEdf[df.NARRATIVE.notnull()]#[df['NARRATIVE'].str.contains("ball")]

# bbls_filtered = df[df['NARRATIVE'].fillna('').str.contains('BBLS')]
# bbbls_filtered = df[df['NARRATIVE'].fillna('').str.contains('BBBLS')]
# # pattern = re.findall(r'\d+ (BBLS|bbls)', data_filtered)
# l = [i for i in bbls_filtered.index]

# df.PRPTY = df.PRPTY.apply(str)
# df.loc[l,'PRPTY' ] += ' BBLS'
```

```python
df['CAUS'][df.CAUS == "OTHER"] = 'other'
df['CAUS'][df.CAUS == "OUTSIDE FORCE DAMAGE"] = 'other'
df['CAUS'][df.CAUS == "CORROSION"] = 'maintenance'
df['CAUS'][df.CAUS == "INCORRECT OPERATION BY OPERATOR PERSONNEL"] = 'technical failure'
df['CAUS'][df.CAUS == "FAILED PIPE"] = 'leaks and spills'
df['CAUS'][df.CAUS == "MALFUNCTION OF CONTROL OR RELIEF EQUIPMENT"] = 'maintenance'
df['CAUS'][df.CAUS == "FAILED WELD"] = 'leaks and spills'


# df['CAUS'][df.CAUS == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df.CAUS.value_counts()
```

```python
df.rename(columns = {'IDATE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'ACSTATE': 'State',
                     'CAUS'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'PRPTY' : 'Estimated total property damage',
                     'LOSS'  : 'Final Extracted Spill Amount (Barrels)'},
                      inplace = True)
```

```python
df.insert(3, column='Name of Installation', value=data.ACCOUNTY.values.astype(str))

df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%Y%m%d')


df['Final Extracted Spill Amount (Barrels)'] = df['Final Extracted Spill Amount (Barrels)'].fillna(0).astype(int)
```

```python
df
```

```python

writer = pd.ExcelWriter('accident_hazardous_liquid_1986_jan2002.xlsx')
df.to_excel(writer,'accident_hazardous_liquid_1986_jan2002')
writer.save()
```

# Hazardous liquid 2010 to present

```python
file3 = "sources_from_links/accident_hazardous_liquid_jan2010_present/accident_hazardous_liquid_jan2010_present.xls"
```

```python
data3 = pd.read_excel(file3)
```

```python
df3 = pd.DataFrame(data=data3, columns=['LOCAL_DATETIME','OPERATOR_STATE_ABBREVIATION', 'CAUSE','NAME', 
                                        'NARRATIVE','INTENTIONAL_RELEASE_BBLS','UNINTENTIONAL_RELEASE_BBLS'])
```

```python
df3.INTENTIONAL_RELEASE_BBLS = df3.INTENTIONAL_RELEASE_BBLS.fillna(0)
```

```python
df3['Final Extracted Spill Amount (Barrels)'] = df3.INTENTIONAL_RELEASE_BBLS + df3.UNINTENTIONAL_RELEASE_BBLS
df3.drop('INTENTIONAL_RELEASE_BBLS',1, inplace=True)
df3.drop('UNINTENTIONAL_RELEASE_BBLS',1,inplace=True)
```

```python
df3['CAUSE'][df3.CAUSE == "EQUIPMENT FAILURE"] = 'maintenance'
df3['CAUSE'][df3.CAUSE == "CORROSION FAILURE"] = 'maintenance'
df3['CAUSE'][df3.CAUSE == "INCORRECT OPERATION"] = 'technical failure'
df3['CAUSE'][df3.CAUSE == "MATERIAL FAILURE OF PIPE OR WELD"] = 'leaks and spills'
df3['CAUSE'][df3.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'
df3['CAUSE'][df3.CAUSE == "OTHER INCIDENT CAUSE"] = 'other'
df3['CAUSE'][df3.CAUSE == "OTHER OUTSIDE FORCE DAMAGE"] = 'other'

df3['CAUSE'][df3.CAUSE == "EXCAVATION DAMAGE"] = 'sabotage'
# df3['CAUSE'][df3.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df3.CAUSE.value_counts()
```

```python
df3.rename(columns = {'LOCAL_DATETIME'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'OPERATOR_STATE_ABBREVIATION': 'State',
                     'CAUSE'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text'},
                      inplace = True)
```

```python
df3.head()
```

```python
df3.insert(3, column='Name of Installation', value=data3.ONSHORE_COUNTY_NAME.values.astype(str))

df3['Name of Installation'] = 'UNKNOWN (' + df3['Name of Installation'] + ' COUNTY )'

# df2['Incident Date'] =  pd.to_datetime(df2['Incident Date'], format='%m/%d/%Y')

df3['Final Extracted Spill Amount (Barrels)'] = df3['Final Extracted Spill Amount (Barrels)'].fillna(0).astype(int)
```

## Gas incident: Jan 2010 - present

```python
file = "sources_from_links/incident_gas_distribution_jan2010_present/incident_gas_distribution_jan2010_present.xls"
```

```python
data = pd.read_excel(file)
```

```python
# temp_df = excel_to_dataframe(file2)
# df2 = pd.DataFrame(data=temp_df, columns=['IDATE','ACSTATE', 'CAUS','NAME', 'NARRATIVE','PRPTY', 'LOSS'])
```

```python
# temp_df.loc[0]
```

```python
df = pd.DataFrame(data=data, columns=['LOCAL_DATETIME','LOCATION_STATE_ABBREVIATION', 'CAUSE','NAME', 'NARRATIVE', 'EST_COST_GAS_RELEASED'])
```

```python
df.head(10)
```

```python
df.CAUSE.value_counts()
```

```python
df['CAUSE'][df.CAUSE == "EQUIPMENT FAILURE"] = 'maintenance'
df['CAUSE'][df.CAUSE == "CORROSION FAILURE"] = 'maintenance'
df['CAUSE'][df.CAUSE == "INCORRECT OPERATION"] = 'technical failure'
df['CAUSE'][df.CAUSE == "MATERIAL FAILURE OF PIPE OR WELD"] = 'leaks and spills'
df['CAUSE'][df.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'
df['CAUSE'][df.CAUSE == "OTHER INCIDENT CAUSE"] = 'other'
df['CAUSE'][df.CAUSE == "OTHER OUTSIDE FORCE DAMAGE"] = 'other'

df['CAUSE'][df.CAUSE == "EXCAVATION DAMAGE"] = 'sabotage'
# df['CAUSE'][df.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df.rename(columns = {'LOCAL_DATETIME'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'LOCATION_STATE_ABBREVIATION': 'State',
                     'CAUSE'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'EST_COST_GAS_RELEASED'  : 'Total Gas Released ($)'},
                      inplace = True)
```

```python
r
df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

# df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%m/%d/%Y')

df['Total Gas Released ($)'] = df['Total Gas Released ($)'].fillna(0).astype(int)
```

```python
df['Sources'] = "https://www.phmsa.dot.gov/pipeline/library/data-stats/distribution-transmission-and-gathering-lng-and-liquid-annual-data"
```

```python
df.head(10)
```

```python
writer = pd.ExcelWriter('incident_gas_distribution_jan_2010_present.xlsx')
df.to_excel(writer,'incident_gas_distribution_jan2010_present')
writer.save()
```

# Gas incident: 2004 to 2009

```python
file2 = "sources_from_links/incident_gas_distribution_mar2004_dec2009/incident_gas_distribution_mar2004_dec2009.xls"
```

```python
data2 = pd.read_excel(file2)
```

```python
df2 = pd.DataFrame(data=data2, columns=['IDATE','ACSTATE', 'CAUSE_TEXT','NAME', 'NARRATIVE', 'GASPRP'])
```

```python
df2
```

```python
df2.CAUSE_TEXT.value_counts()
```

```python
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "EXCAVATION DAMAGE"] = 'sabotage'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "OTHER OUTSIDE FORCE DAMAGE"] = 'other'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "OTHER"] = 'other'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "NATURAL FORCES"] = 'natural disasters'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "MATERIAL AND/OR WELD FAILURES"] = 'leaks and spills'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "INCORRECT OPERATION"] = 'technical failure'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "EQUIPMENT"] = 'maintenance'
df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "CORROSION"] = 'maintenance'

# df2['CAUSE_TEXT'][df2.CAUSE_TEXT == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df2.CAUSE_TEXT.value_counts()
```

```python
df2.rename(columns = {'IDATE'     :'Incident Date',
                     'NAME'       : 'Name of Owner',
                     'ACSTATE'    : 'State',
                     'CAUSE_TEXT' : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'GASPRP'      : 'Total Gas Loss ($)'},
                      inplace = True)
```

```python
df2.head()
```

```python
df2.insert(3, column='Name of Installation', value=data2.ACCOUNTY.values.astype(str))

df2['Name of Installation'] = 'UNKNOWN (' + df2['Name of Installation'] + ' COUNTY )'

# df2['Incident Date'] =  pd.to_datetime(df2['Incident Date'], format='%m/%d/%Y')

df2['Total Gas Loss ($)'] = df2['Total Gas Loss ($)'].fillna(0).astype(int)
```

```python
df2.head()
```

```python
writer = pd.ExcelWriter('incident_gas_distribution_mar_2004_dec_2009.xlsx')
df2.to_excel(writer,'incident_gas_distribution_mar2004_dec2009')
writer.save()
```

## Gas incident: 1984 - 2004

```python
file3 = "sources_from_links/incident_gas_distribution_mid1984_feb2004/incident_gas_distribution_mid1984_feb2004.xls"
```

```python
data3 = pd.read_excel(file3)
```

```python
data3.head()
```

```python
df3 = pd.DataFrame(data=data3, columns=['IDATE','ACCST', 'CAUSE', 'MANU','NAME', 'NARRATIVE', 'PRPTY'])
```

```python
df3.head(10)
```

```python
df3.MANU = df3.MANU.fillna("").astype(str)
indice = df3['MANU'][df3.MANU == ''].index
for i in indice:
    df3['MANU'][i] = 'UNKNOWN (' + data3.ACCNT.values.astype(str)[i] + ' COUNTY)'
```

```python
df3.head(10)
```

```python
df3.CAUSE.value_counts()
```

```python
df3['CAUSE'][df3.CAUSE == "DAMAGE BY OUTSIDE FORCES"] = 'other'
df3['CAUSE'][df3.CAUSE == "OTHER"] = 'other'
df3['CAUSE'][df3.CAUSE == "CONSTRUCTION/OPERATING ERROR"] = 'technical failure'
df3['CAUSE'][df3.CAUSE == "CORROSION"] = 'maintenance'
df3['CAUSE'][df3.CAUSE == "ACCIDENTALLY CAUSED BY OPERATOR "] = 'sabotage'
# df3['CAUSE'][df3.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df3.rename(columns = {'IDATE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'ACCST': 'State',
                     'CAUSE'   : 'Cause of Loss', 
                      'MANU'  : 'Name of Installation',
                     'NARRATIVE'  : 'Text',
                     'PRPTY'  : 'Total Gas Released ($)'},
                      inplace = True)
```

```python
# df3.insert(3, column='Name of Installation', value=data3.MANU.values.astype(str))

# df3['Name of Installation'][df3['Name of Installation'] == ''] = 'UNKNOWN (' + data3.ACCNT.values.astype(str) + ' COUNTY)'

df3['Incident Date'] =  pd.to_datetime(df3['Incident Date'], format='%Y%m%d')
# df3['Name of Installation'] = df3['Name of Installation'].fillna('UNKNOWN').astype(str)
df3['Total Gas Released ($)'] = df3['Total Gas Released ($)'].fillna(0).astype(int)
```

```python
df3['Sources'] = "https://www.phmsa.dot.gov/pipeline/library/data-stats/distribution-transmission-and-gathering-lng-and-liquid-annual-data"
```

```python
df3.head(10)
```

```python
writer = pd.ExcelWriter('incident_gas_distribution_mid_1984_feb_2004.xlsx')
df3.to_excel(writer,'incident_gas_distribution_mid1984_feb2004')
writer.save()
```

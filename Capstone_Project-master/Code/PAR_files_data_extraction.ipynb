```python
def convert(fname, pages=None):
    if not pages:
        pagenums = set()
    else:
        pagenums = set(pages)

    output = StringIO()
    manager = PDFResourceManager()
    converter = TextConverter(manager, output, laparams=LAParams())
    interpreter = PDFPageInterpreter(manager, converter)

    infile = file(fname, 'rb:pdf')
    for page in PDFPage.get_pages(infile, pagenums):
        interpreter.process_page(page)
    infile.close()
    converter.close()
    text = output.getvalue()
    output.close
    return text
```

```python
def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = file(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
import glob

text_lst = []
for filename in glob.glob('PAR/*.pdf'):
    try: 
        text_lst.append(convert_pdf_to_txt(filename))
        print("Successful read file: ", filename)
    except:
        print("Can't read file: ", filename)
```

```python
def extract_data(text_lst):
    texts = []
    dates = []
    states = []
    losses = []
    causes = []
    owners = []
    deaths = []
    injuries = []
    for i, textfile in enumerate(text_lst):
        try:
            print(i)
            #get accident description
            accident_detail = re.sub(r'\n', "@", textfile)
            accident_detail = re.findall(r'(?<=Abstract: ).*?@@', accident_detail)

            #get date time
            date_and_time = re.findall(r'.*,\s+\d{4}', accident_detail[0])      
            date = re.findall(r'[A-Z]\w+\s+\d+[,\s]+\d{4}', str(date_and_time))
            time = re.findall(r'[\s\d]\d:\d\d\s+[ap]\.m\.', str(date_and_time))
            date_time =  str(date) + " " + str(time)

            #get State
            state = re.findall(r',\s+[A-Z]\w+[\.,]', accident_detail[0])
            state = re.sub(r', |\.|,', "", state[0])

            #get loss
            loss = re.findall(r'(?<=\$).*', accident_detail[0])
            
            #get cause of loss
            cause = re.findall(r'ruptur|leak|pipe failure|explod|fire|puncture|spill|material defect',accident_detail[0])
            
            #get owner/operator
            owner = re.findall(r'(?<=operat).*?,|(?<=own).*?\,', accident_detail[0])
            
            #get death and injuries
            death = re.findall(r'\.\s+[A-Z]\w+.*killed|\.\s+[A-Z]\w+.*died|(?<=\.).*death', accident_detail[0])
            injury = re.findall(r'\.\s+[A-Z]\w+.*injured|\.\s+[A-Z]\w+.*injuries', accident_detail[0])
            
            if accident_detail != []:
                texts.append(accident_detail[0])
                dates.append(date_time)
                states.append(state)
                losses.append(loss)
                causes.append(cause)
                owners.append(owner)
                deaths.append(death)
                injuries.append(injury)
        except:
            print("Problem at file:", i)


        
    return texts, dates, states, losses, causes, owners, deaths, injuries
```

```python
def text2int(textnum, numwords={}):
    if not numwords:
      units = [
        "zero", "one", "two", "three", "four", "five", "six", "seven", "eight",
        "nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen",
        "sixteen", "seventeen", "eighteen", "nineteen",
      ]

      tens = ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

      scales = ["hundred", "thousand", "million", "billion", "trillion"]

      numwords["and"] = (1, 0)
      for idx, word in enumerate(units):    numwords[word] = (1, idx)
      for idx, word in enumerate(tens):     numwords[word] = (1, idx * 10)
      for idx, word in enumerate(scales):   numwords[word] = (10 ** (idx * 3 or 2), 0)

    current = result = 0
    word_lst = []
    for word in textnum.split():
        if word not in numwords:
#           raise Exception("Illegal word: " + word)
#             continue
            word_lst.append(word)
        else:
            scale, increment = numwords[word]
            current = current * scale + increment
            if scale > 100:
                result += current
                current = 0

    return word_lst, result + current

print text2int("seven billion one hundred million thirty one thousand three hundred thirty seven")
#7100031337
```

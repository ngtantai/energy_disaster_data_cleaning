```python
def page_iteration(webpages):
    """return all incident events of each page"""
    incident_pages = []
    states = []
    dates = []
    for i,webpage in enumerate(webpages):
        #Query the website and return the html to the variable 'page'
        page = urllib.request.urlopen(webpage)

        #Parse the html in the 'page' variable, and store it in Beautiful Soup format
        soup = BeautifulSoup(page, 'lxml')

        #get all links appear on page
        all_links = soup.find_all('a', class_="name")
        all_locations = soup.find_all('span', class_="location")
        all_dates = soup.find_all('td', class_="date")
        prefix = "https://incidentnews.noaa.gov"
        
        for j,link in enumerate(all_links):
            incident_pages.append(prefix + link.get("href"))
            states.append(re.findall(r'[A-Z]{2}(?=\s)', str(all_locations[j])))
            dates.append(re.findall(r'(?<=\>).*(?=\<)', str(all_dates[j])))
        


    return incident_pages, states, dates
```

```python
incident_pages, states, dates = page_iteration(webpages)
```

```python
def extract_info(page):
    """extract metric names and metric value on a single page"""
    soup = BeautifulSoup(page, "lxml")
    texts = soup.find_all('p', class_="sub")
    info = soup.find_all('td')
   
    texts = re.findall(r'(?<=strong\>\:).*', str(texts))
#     texts = re.sub(r'\]|\[|\\', "", str(texts))
#     print(texts)
    #clean info
    info = re.sub(r'\<.*?\>',"",str(info))
    info = re.sub(r'\:\s*,', ":", str(info))
    info = re.sub(r'\[|\]', "", str(info))
    
    #get metrics
    metrics = re.findall(r'(?<=,\s)[A-Z]\w+[^,]*?\:|^.*?\:', str(info))
    metrics = [item.replace(":","") for item in metrics]
    metrics.append('Text')
    
    #get values
    values = re.findall(r'(?<=\:\s).*?(?=,)|(?<=\:\s).*?(?=$)', str(info))
    values.append(texts)
    same_size = False
    if (len(metrics) == len(values)):
        same_size = True        
    return metrics, values, same_size
```

```python
def multi_page_data_extraction(incident_pages):
    """return a dataframe contain all data on a single page"""
    dfs = []
    size_unmatch = []
    for i,webpage in enumerate(incident_pages):
        page = urllib.request.urlopen(webpage)
        metrics, values, same_size = extract_info(page)
        if same_size == True:
            df = pd.DataFrame(data=[values], columns=metrics)
            dfs.append(df)
        else:
            size_unmatch.append(i)
    return dfs, size_unmatch
```

```python
dfs, size_unmatch = multi_page_data_extraction(incident_pages)
df = pd.concat(dfs, axis=0).reset_index().drop(['index'], axis=1)
```

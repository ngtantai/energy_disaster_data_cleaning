```python
f1 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_1990.txt"
f2 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_1991.txt"
f3 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_1992.txt"
f4 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_1993.txt"
f5 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_1994.txt"
f6 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_2000.txt"
f7 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_2001.txt"
f8 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_2002.txt"
f9 = "sources_from_links/annual_gas_distribution_1970_2003/annual_gas_distribution_2003.txt"
file =[f1,f2,f3,f4,f5,f6,f7,f8,f9]
```

```python
file_object  = open(“filename”, “mode”) 
```

```python
data = pd.read_table(file[0])
```

```python
data
```

```python
df = pd.DataFrame(data=data, columns=[ 'DOE','OSTATE', 'CAUS', 'OCOUNTY', 'NAME', 'COMMENTS',
                                       'CM', 'CS', 'TM', 'TS', 'OM', 'OS', 'CDM', 'CDS',
                                       'MDM', 'MDS', 'OTHM', 'OTHS', 'KNLK', 'FEDLK'])
```

```python
df
```

```python
df['Final Extracted Spill Amount (Barrels)'] = df.CM + df.CS + df.TM + df.TS + df.OM + \
                                                df.OS + df.CDM + df.CDS + df.MDM + df.MDS + df.OTHM + df.OTHS +\
                                                df.KNLK + df.FEDLK
```

```python
df
```

```python
df = df[['DOE','OSTATE', 'CAUS', 'OCOUNTY', 'NAME', 'COMMENTS','Final Extracted Spill Amount (Barrels)']]
df.rename(columns = {'DOE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'OSTATE': 'State',
                     'CAUS'   : 'Cause of Loss', 
                     'COMMENTS'  : 'Text',
                     'OCOUNTY' : 'Name of Installation'},
                      inplace = True)
```

```python
df
```

```python
# df.insert(3, column='Name of Installation', value=data.COUNT.values.astype(str))

df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%m/%d/%Y')


df['Final Extracted Spill Amount (Barrels)'] = df['Final Extracted Spill Amount (Barrels)'].fillna(0).astype(int)
df['Cause of Loss'] = df['Cause of Loss'].fillna('').astype(str)
df['Text'] = df['Text'].fillna('').astype(str)
```

```python
# # df.NARRATIVE.values.any
# empty_list = []
# df.NARRATIVE.fillna('').head()
# # df.NARRATIVEdf[df.NARRATIVE.notnull()]#[df['NARRATIVE'].str.contains("ball")]

# bbls_filtered = df[df['NARRATIVE'].fillna('').str.contains('BBLS')]
# bbbls_filtered = df[df['NARRATIVE'].fillna('').str.contains('BBBLS')]
# # pattern = re.findall(r'\d+ (BBLS|bbls)', data_filtered)
# l = [i for i in bbls_filtered.index]

# df.PRPTY = df.PRPTY.apply(str)
# df.loc[l,'PRPTY' ] += ' BBLS'
```

```python
df['CAUS'][df.CAUS == "OTHER"] = 'other'
df['CAUS'][df.CAUS == "OUTSIDE FORCE DAMAGE"] = 'other'
df['CAUS'][df.CAUS == "CORROSION"] = 'maintenance'
df['CAUS'][df.CAUS == "INCORRECT OPERATION BY OPERATOR PERSONNEL"] = 'technical failure'
df['CAUS'][df.CAUS == "FAILED PIPE"] = 'leaks and spills'
df['CAUS'][df.CAUS == "MALFUNCTION OF CONTROL OR RELIEF EQUIPMENT"] = 'maintenance'
df['CAUS'][df.CAUS == "FAILED WELD"] = 'leaks and spills'


# df['CAUS'][df.CAUS == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df.CAUS.value_counts()
```

```python
df.rename(columns = {'IDATE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'ACSTATE': 'State',
                     'CAUS'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'PRPTY' : 'Estimated total property damage',
                     'LOSS'  : 'Final Extracted Spill Amount (Barrels)'},
                      inplace = True)
```

```python
df.insert(3, column='Name of Installation', value=data.ACCOUNTY.values.astype(str))

df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%Y%m%d')


df['Final Extracted Spill Amount (Barrels)'] = df['Final Extracted Spill Amount (Barrels)'].fillna(0).astype(int)
```

```python
df
```

```python

writer = pd.ExcelWriter('accident_hazardous_liquid_1986_jan2002.xlsx')
df.to_excel(writer,'accident_hazardous_liquid_1986_jan2002')
writer.save()
```

## Gas Distribution: 1986 - feb 2004

```python
file = "sources_from_links/PHMSA_Pipeline_Safety_Flagged_Incidents/gd1986tofeb2004.xlsx"
```

```python
data = pd.read_excel(file, sheetname="gd1986tofeb2004")
```

```python
data.head()
```

```python
df = pd.DataFrame(data=data, columns=['IDATE','ACCST', 'CAUSE',
                                      'NAME', 'NARRATIVE', 'TOTAL_COST'])
```

```python
df.head(10)
```

```python
df.CAUSE.value_counts()
```

```python
df['CAUSE'][df.CAUSE == "DAMAGE BY OUTSIDE FORCES"] = 'other'
df['CAUSE'][df.CAUSE == "OTHER"] = 'other'
df['CAUSE'][df.CAUSE == "CONSTRUCTION/OPERATING ERROR"] = 'technical failure'
df['CAUSE'][df.CAUSE == "ACCIDENTALLY CAUSED BY OPERATOR"] = 'sabotage'
df['CAUSE'][df.CAUSE == "NO DATA"] = ''
df['CAUSE'][df.CAUSE == "CORROSION"] = 'maintenance'
# df['CAUSE'][df.CAUSE == "MATERIAL AND/OR WELD FAILURES"] = 'leaks and spills'
# df['CAUSE'][df.CAUSE == "NATURAL FORCES"] = 'natural disasters'
# df['CAUSE'][df.CAUSE == "EQUIPMENT"] = 'maintenance'


# df['CAUSE'][df.CAUSE == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df.rename(columns = {'IDATE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'ACCST': 'State',
                     'CAUSE'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'TOTAL_COST'  : 'Total Gas Released ($)'},
                      inplace = True)
```

```python
df.insert(3, column='Name of Installation', value=data.ACCNT.values.astype(str))

df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

# df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%m/%d/%Y')

df['Total Gas Released ($)'] = df['Total Gas Released ($)'].fillna('').astype(str)
```

```python
df['Sources'] = "https://www.phmsa.dot.gov/pipeline/library/data-stats/flagged-data-files"
```

```python
df.head(10)
```

```python
writer = pd.ExcelWriter('gas_distribution_1986_to_feb_2004.xlsx')
df.to_excel(writer,'gd1986tofeb2004')
writer.save()
```

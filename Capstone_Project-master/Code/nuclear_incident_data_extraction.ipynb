```python
import pandas as pd
import numpy as np
import re
import pycountry
from textblob import TextBlob
from calendar import month_name
from collections import Counter
```

```python
file = 'nuclear_indcident_1952_2011.xlsx'
data = pd.read_excel(file)
```

```python
# df = pd.DataFrame(data=data, columns=['Year', 'Cause of Loss', 'Deaths', 'Loss Estimated ($)', 'Location', 'Text'])
df = pd.DataFrame(data=data)
df
```

```python
df.Location = df.Location.apply(lambda x: str(x).replace(',\xa0','-'))
```

```python
df.Location = df.Location.apply(lambda x: re.sub(r'[A-Z]\w+\s?\w+[\s,]',"", str(x)))
```

```python
df.Location[4]
```

```python
df.head(10)
```

```python
df['Country'] = df.Location.apply(lambda x: re.findall(r'(?<=\-)\w+$', str(x)))
df['Country'] = df.Country.apply(lambda x: re.sub(r'\[|\]|\'', "", str(x)))
```

```python
df.head(10)
```

```python
df['States'] = df.Location.apply(lambda x: re.findall(r'(?=\w+\-)\w+', str(x)))
df['States'] = df.States.apply(lambda x: re.sub(r'\[|\]|\'|,', "", str(x)))
```

```python
df
```

```python
df = pd.DataFrame(data=df, columns=['Date','Country','States','Deaths','Loss estimated ($ millions)', 'Text'])
```

```python
df.sort_values(['Date'], inplace=True)
```

```python
writer = pd.ExcelWriter('nuclear_indcident_1905_2011.xlsx')
df.to_excel(writer,'nuclear_indcident1905to2011')
writer.save()
```

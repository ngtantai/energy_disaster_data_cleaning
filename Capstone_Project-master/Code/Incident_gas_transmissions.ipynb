## Gas transmission incident: 2002 - 2009

```python
file = "sources_from_links/incident_gas_transmission_gathering_2002_dec2009/incident_gas_transmission_gathering_2002_dec2009.xls"
```

```python
data = pd.read_excel(file)
```

```python
data.head()
```

```python
df = pd.DataFrame(data=data, columns=['IDATE','ACSTATE', 'CAUSE_TEXT',
                                      'NAME', 'NARRATIVE', 'PRPTY'])
```

```python
df.head(10)
```

```python
df.CAUSE_TEXT.value_counts()
```

```python
df['CAUSE_TEXT'][df.CAUSE_TEXT == "CORROSION"] = 'maintenance'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "MATERIAL AND/OR WELD FAILURES"] = 'leaks and spills'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "EXCAVATION"] = 'sabotage'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "NATURAL FORCES"] = 'natural disasters'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "OTHER"] = 'other'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "OTHER OUTSIDE FORCE DAMAGE"] = 'other'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "EQUIPMENT"] = 'maintenance'
df['CAUSE_TEXT'][df.CAUSE_TEXT == "INCORRECT OPERATION"] = 'technical failure'


# df['CAUSE_TEXT'][df.CAUSE_TEXT == "NATURAL FORCE DAMAGE"] = 'natural disasters'

```

```python
df.rename(columns = {'IDATE'  :'Incident Date',
                     'NAME'   : 'Name of Owner',
                     'ACSTATE': 'State',
                     'CAUSE_TEXT'   : 'Cause of Loss', 
                     'NARRATIVE'  : 'Text',
                     'PRPTY'  : 'Total Gas Released ($)'},
                      inplace = True)
```

```python
df.insert(3, column='Name of Installation', value=data.ACCOUNTY.values.astype(str))

df['Name of Installation'] = 'UNKNOWN (' + df['Name of Installation'] + ' COUNTY)'

# df['Incident Date'] =  pd.to_datetime(df['Incident Date'], format='%m/%d/%Y')

df['Total Gas Released ($)'] = df['Total Gas Released ($)'].fillna(0).astype(int)
```

```python
df['Sources'] = "https://www.phmsa.dot.gov/pipeline/library/data-stats/distribution-transmission-and-gathering-lng-and-liquid-annual-data"
```

```python
df.head(10)
```

```python
writer = pd.ExcelWriter('incident_gas_transmission_gathering_2002_dec_2009.xlsx')
df.to_excel(writer,'incident_gas_transmission_gathering_2002_dec2009')
writer.save()
```

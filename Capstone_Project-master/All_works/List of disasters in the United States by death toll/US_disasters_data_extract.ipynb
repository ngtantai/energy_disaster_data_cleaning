```python
import pandas as pd
import numpy as np
import re
import pycountry
from textblob import TextBlob
from calendar import month_name
from collections import Counter
```

```python
file = 'USA_disasters_by_death_toll.xlsx'
data = pd.read_excel(file)
```

```python
# df = pd.DataFrame(data=data, columns=['Year', 'Cause of Loss', 'Deaths', 'Loss Estimated ($)', 'Location', 'Text'])
df = pd.DataFrame(data=data)
df
```

```python
df = df.dropna(axis=0, how='all', subset=['Year'])
```

```python
df.Year = df.Year.astype(int)
```

```python
df.head()
```

```python
df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: re.findall(r'\d.*0',str(x)))
df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: str(x).replace(',',"").replace('[',"")
                                                                             .replace(']', "")
                                                                             .replace('\'',""))
# df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: re.findall(r'^\[\d+0',str(x)))
```

```python
df
```

```python
# df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: str(x).replace('[',"")
#                                                                              .replace(']', "")
#                                                                              .replace('\'',"").replace('(',""))
df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: re.findall(r'^\d+0',str(x)))
df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: str(x).replace(',',"").replace('[',"")
                                                                             .replace(']', "")
                                                                             .replace('\'',""))
```

```python
df
```

```python
df['Loss Estimated (US$)'] = df['Loss Estimated (US$)'].apply(lambda x: int(x) if str(x) != '' else x)
```

```python
# df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: str(x).replace('$',"")
#                                                                              .replace('(approximate)', "")
#                                                                              .replace(',',"").replace('<',"").replace('>',"")
#                                                                              .replace('-',"").replace('nan',"0").replace("~",""))
# df['Loss Estimated (US$)'] =df['Loss Estimated (US$)'].apply(lambda x: int(x))
```

```python
df['Location'] =df['Location'].apply(lambda x: re.sub(r'\(.*\)', "",str(x)))

# df['Location'] =df['Location'].apply(lambda x: re.findall(r'[A-Z]\w+?\s?\w+$',str(x)))
# df['Location'] =df['Location'].apply(lambda x: re.findall(r'[A-Z]?\w+?\s?[A-Z]?\w+$',str(x)))
```

```python
df['Location'] =df['Location'].apply(lambda x: re.findall(r'[A-Z]?\w+?\s?\w+$',str(x)))
df['Location'] =df['Location'].apply(lambda x: str(x).replace('[',"").replace(']', "").replace("\'",""))
                                                                     
```

```python
df
```

```python
df.drop(labels='Cause of Loss.1',axis=1, inplace=True)
```

```python
df.insert(1, 'Country', value='USA')
```

```python
df['Source'] = 'https://en.wikipedia.org/wiki/List_of_disasters_in_the_United_States_by_death_toll'
```

```python
df['Year'] = df['Year'].apply(lambda x: int(x))
```

```python
df.sort_values(['Year'], inplace=True)
```

```python
import pandas as pd
import re
import pycountry


file = "oil_spills.xlsx"
data = pd.read_excel(file)
data
```

```python
data['Countries'] = data['Location']
data['Countries'] = data['Countries'].apply(lambda x: str(x).replace(str(x)[0],''))
data['Countries'] = data['Countries'].apply(lambda x: re.findall(r'^[A-Z]\w+\s\w+|^[A-Z]\w+,', x))
data['Countries'] = data['Countries'].apply(lambda x: str(x).replace(',','')
                                          .replace('[','').replace(']','').replace('\'',""))
```

```python
data['County'] = data['Location'] 
data['County'] = data['County'].apply(lambda x: re.findall(r'[A-Z]\w+\sCounty', str(x)))
data['County'] = data['County'].apply(lambda x: str(x).replace(',','')
                                          .replace('[','').replace(']','').replace('\'',""))
data['County'] = data['County'].fillna('UNKNOWN').astype(str)
```

## Clean date

```python
data['Dates'] = data['Dates'].apply(lambda x: re.findall(r'\d{4}-\d{2}-\d{2}', str(x))[0] if len(str(x))>20 else str(x))
data['Dates'] = data['Dates'].fillna('UNKNOWN').replace("nan","UNKNOWN")
```

```python
data = data.fillna('UNKNOWN').replace("nan","UNKNOWN")
```

```python
data
```

## Clean Min/Max spilled oil

```python
data['Min Tonnes'] = data['Min Tonnes'].apply(lambda x: re.findall(r'(?<=♠)\d+\.?\d+', str(x)))
data['Max Tonnes'] = data['Max Tonnes'].apply(lambda x: re.findall(r'(?<=♠)\d+\.?\d+', str(x)))
```

```python
data['Min Tonnes'] = data['Min Tonnes'].apply(lambda x: str(x).replace('[','').replace(']','').replace('\'',""))
data['Max Tonnes'] = data['Max Tonnes'].apply(lambda x: str(x).replace('[','').replace(']','').replace('\'',""))
```

```python
df = pd.DataFrame(data=data, columns=['Dates', 'Countries', 'County', 'Spill / Vessel', 'Max Tonnes', 'Link(s)'])
```

```python
df['County'] = df['County'].replace("",'UNKNOWN')
```

```python
df['Max Tonnes'] = round(df['Max Tonnes'].apply(pd.to_numeric) * 264.17 / 42, 0)
```

```python
df = df.rename(index=str, columns={"Link(s)": "Sources", 
                                   "County": "Name of Installation",
                                   "Max Tonnes": "Spill amount (barrels)", 
                                   "Spill / Vessel": "Name of Owner"})
df['Sources'] = "https://en.wikipedia.org/wiki/List_of_oil_spills#cite_note-:0-192"
```

```python
df['Spill amount (barrels)'] = df['Spill amount (barrels)'].fillna('Unknown')
```

```python
df['Name of Owner'] = df['Name of Owner'].apply(lambda x: re.sub(r'([Oo]il|[Ss][Pp]ill|[Ll]eak|[Pp]ipeline)',"", str(x)))
```

```python
def find_county(alist):
    counties = []
    for loc in alist:
        tokens = loc.replace(',',"").split(" ")
        a = 'unknown'
        for i,v in enumerate(tokens):
            if v == 'County':
                a = tokens[i-1]+' '+v
        counties.append(a)
    return counties
```

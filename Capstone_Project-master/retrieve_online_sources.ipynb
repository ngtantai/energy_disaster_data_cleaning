## Top 10 recent American industrial disasters

## Refinery accidents explosions 

```python
#specify the url
refinery_accidents_explosions = "https://ehssafetynewsamerica.com/category/refinery-accidents-explosions/"

#Query the website and return the html to the variable 'page'
page = urllib.request.urlopen(refinery_accidents_explosions)

#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page)

web_content = soup.prettify()
print(web_content)

```

### Get into subpages

```python
main_links = soup.find_all("a", rel="bookmark")
list_of_links = []
for link in main_links:
    href = str(link.get("href"))
    text = re.findall(r'https:\/\/\w+.com\/\d+\/\d+\/\d+\/.+\/', href)
    if text != []:
        list_of_links.append(text[0])
```

```python
for i,v1 in enumerate(list_of_links):
    for j,v2 in enumerate(list_of_links):
        if v1 == v2:
            list_of_links.remove(v2)
```

### Compile subpages

```python
for link in list_of_links:

    #Query the website and return the html to the variable 'page'
    page = urllib.request.urlopen(link)
    print(link)
    #Parse the html in the 'page' variable, and store it in Beautiful Soup format
    soup = BeautifulSoup(page)

    web_content = soup.prettify()
#     print(web_content)
    
    name = re.findall(r'\w+\-.+[^\/]', str(link))
 

    if name != []:
        html_file = open(str(name[0]),"w")
        html_file.write(web_content)
        html_file.close()
```
